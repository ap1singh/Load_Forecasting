## Plot predictions with and without noise

```python
for region in regions:
    for energy in energies:
        for gdp_type in gdp_types:
            plt.plot(with_noise[region][energy][gdp_type], label = 'with noise')
            plt.plot(predictions[energy][gdp_type][region], label= 'no noise')
            plt.legend()
            title = f'{region}_{energy}_{gdp_names[gdp_type]}-GDP-growth'
            plt.title(title)
            plt.show()
            plt.savefig(f'images/{title}.png')
```
