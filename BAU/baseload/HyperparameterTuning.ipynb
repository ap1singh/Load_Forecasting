```python
from config import future_yrs, historical_yrs, cities as regions, energies, gdp_types
future_yrs = [2000 + yr for yr in future_yrs]
historical_yrs = [2000 + yr for yr in historical_yrs]
```

```python
from predict import load_data, initialize_models

energies, past_Xs, future_Xs = load_data()
peak = energies['peak']
consumption = energies['consumption']
```

```python
models = {}
avg_nlls = {}
for i in range(20):
    l1 = (.05*i+.05)
    models[i] = initialize_models(energies, past_Xs, future_Xs, l1_ratio=l1)
    avg_nlls[l1] = sum(model.metrics()[1] for model in models[i])/10.0
```

```python

avg_nlls
```

```python
lists = sorted(avg_nlls.items()) # sorted by key, return a list of tuples

x, y = zip(*lists) # unpack a list of pairs into two tuples
plt.plot(x, y)
plt.xlabel('L1 Ratio')
plt.ylabel('Negative Log Likelihood')
plt.show()
```

```python
models = {}
avg_nlls_a = {}
for i in range(3):
    a = 1.0/(i+1)
    models[a] = initialize_models(energies, past_Xs, future_Xs, l1_ratio=.8, alpha=a)
    avg_nlls_a[a] = sum(model.metrics(silent=True, test_yr=2018)[1] for model in models[a])/10.0
print(avg_nlls_a)
lists = sorted(avg_nlls_a.items()) # sorted by key, return a list of tuples

x, y = zip(*lists) # unpack a list of pairs into two tuples
plt.plot(x, y)
plt.show()
```
